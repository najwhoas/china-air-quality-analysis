#### Pemeriksaan outlier pada data

```python
for data in df.columns:
  if df[data].dtype in ['int64', 'float64']:
    q25, q75 = np.percentile(df[data], 25), np.percentile(df[data], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in df[data] if x < minimum or x > maximum]

    print("Outliers pada {}: {}".format(data, len(outliers)))
  else:
    None
```

Tidak terdeteksi adanya outlier pada dataset.

### Explore Partikulat 2.5

```python
partikulat_pivot = df.groupby('Wilayah')[['Partikulat_2.5','Partikulat_10',]].sum()
print(partikulat_pivot)
```

Sementara, daerah dengan jumlah penyebaran Partikulat 10 terbanyak pada data adalah Gucheng.

Berdasarkan informasi pada bmkg.go.id, Particulate Matter (PM2.5) adalah partikel udara yang berukuran lebih kecil dari atau sama dengan 2.5 µm (mikrometer). Karena ukurannya yang super kecil, partikel halus ini tak dapat disaring oleh tubuh. Singkat kata, partikel ini dapat masuk ke tubuh melalui paru-paru dan pembuluh darah.

Mari lihat apakah besar curah hujan dapat memengaruhi jumlah partikel ini dengan korelasi Pearson.

```python
x = df['Curah_Hujan']
y = df['Partikulat_2.5']
korelasi_pm25 = np.corrcoef(x,y)
print("Korelasi Curah Hujan dan Partikulat 2.5: {}".format(korelasi_pm25))
```

Hasil menunjukkan bahwa baik Partikulat 2.5 tidak memiliki hubungan dengan curah hujan.
Mari lihat korelasinya dengan titik embun.

```python
x = df['Titik_Embun']
y = df['Partikulat_2.5']
korelasi_pm25 = np.corrcoef(x,y)
print("Korelasi Titik Embun dan Partikulat 2.5: {}".format(korelasi_pm25))
```

Hasil menunjukkan bahwa besar titik embun memiliki hubungan positif namun lemah dengan Partikulat 2.5

### Bagaimana proporsi polutan SO2, NO2, dan CO pada Tahun 2017 di Wilayah Gucheng?

*gathering data khusus wilayah Gucheng

```python
gucheng_df = pd.read_csv("https://github.com/marceloreis/HTI/raw/master/PRSA_Data_20130301-20170228/PRSA_Data_Gucheng_20130301-20170228.csv")

gucheng_df.fillna(method='ffill', inplace=True)
Q1 = gucheng_df.quantile(0.25)
Q3 = gucheng_df.quantile(0.75)
IQR = Q3 - Q1
maximum = Q3 + (1.5 * IQR)
minimum = Q1 - (1.5 * IQR)
kondisi_lower_than = gucheng_df < minimum
kondisi_more_than = gucheng_df > maximum
dongsi_df = gucheng_df.mask(cond=kondisi_more_than, other=maximum, axis=1)
dongsi_df = gucheng_df.mask(cond=kondisi_lower_than, other=minimum, axis=1)
dongsi_df.info()
```

```python
mean = gucheng_df[gucheng_df['year'] == 2017][['PM2.5','PM10','SO2', 'NO2', 'CO']].mean()
total = mean.sum()

plt.pie( mean/total, autopct='%0.1f%%', colors=['#F39233','#EEF296','#EF4F4F','#9ADE7B','#2D9596'],startangle=140)
plt.axis('equal')
plt.legend(['PM2.5','PM10','SO2', 'NO2', 'CO'])
plt.title('Proporsi Polutan Udara di Wilayah Gucheng Tahun 2017')
plt.show()
```

### Bagaimana tren persebaran polutan Partikulat 2.5, Partikulat 10, Sulfur Dioksida, Nitrogen Dioksida, dan Karbon Monoksida per 6 bulan di wilayah Gucheng?

```python
gucheng_df['waktu'] = df['Waktu_Pengamatan']
data = gucheng_df[['waktu', 'PM2.5', 'PM10', 'SO2','NO2','CO']].set_index('waktu').resample('M').mean()

plt.figure(figsize=(15, 6))
plt.plot(data['PM2.5'], label='Partikulat 2.5', color='#F39233')
plt.plot(data['PM10'], label='Partikulat 10', color='#EEF296')
plt.plot(data['SO2'], label='Sulfur Dioksida', color='#EF4F4F')
plt.plot(data['NO2'], label='Nitrogen Dioksida', color='#9ADE7B')
plt.plot(data['CO'], label='Karbon Dioksida', color='#2D9596')

plt.title('Tren Persebaran Polutan per 6 Bulan di Wilayah Gucheng')
plt.xlabel('Waktu Pengamatan')
plt.ylabel('Kadar Polutan')
plt.legend()
plt.show()
```

### Bagaimana tren keadaan gas ozon dan temperatur udara per 6 Bulan di wilayah Gucheng?

```python
data = gucheng_df[['waktu', 'O3', 'TEMP']].set_index('waktu').resample('M').mean()

plt.figure(figsize=(15, 6))
plt.plot(data['O3'], label='Gas Ozon', color='#F39233')
plt.plot(data['TEMP'], label='Temperatur Udara', color='#2D9596')

plt.xlabel("Waktu Pengamatan")
plt.ylabel("Kadar")

plt.title("Tren Keadaan Gas Ozon dan Temperatur Udara per 6 Bulan di Wilayah Gucheng")

plt.legend()
plt.show()
```

Ini mengindikasikan bahwa polutan berupa gas buang dari kendaraan atau macam-macam alat pembakaran lainnya mendominasi polusi udara di Gucheng.

<br>

> Tren persebaran polutan per 6 bulan di Gucheng menunjukkan bahwa persebaran karbon monoksida selalu mengalami kenaikan yang pesat pada awal tahun. Sementara polutan lainnya cenderung stagnan, tidak terlalu signifikan kenaikan dan penurunannya.

Bagaimana tren keadaan gas ozon dan temperatur udara per 6 Bulan di wilayah Gucheng?

> Berbanding dengan kondisi karbon monoksida, gas ozon selalu mencapai titik tertingginya di pertengahan tahun (Juli). Keadaan temperatur udara juga mengalami pergerakan naik turun yang sama dengan gas ozon. Dengan keadaan ini, dapat diasumsikan bahwa jika temperatur udara sedang tinggi, besar kemungkinan kadar gas O3 juga tinggi. Ini bisa menjadi petunjuk bagi warga Gucheng untuk berantisipasi melindungi diri, seperti menggunakan masker.


<br>

Berdasarkan penelitian, Karbon Monoksida (CO) merupakan salah satu prekursor pembentuk gas O3 yang dihasilkan dari proses fotokimia. Berdasarkan data yang ada, keadaan CO dan O3 mengalami pergerakan berlawanan yang lumayan konsisten. Saat CO berada di titik tinggi, O3 berada di titik rendah.

Dengan keadaan tersebut, dapat diasumsikan bahwa proses fotokimia pembentukan ozon permukaan (O3) terjadi selama kurang lebih 6 bulan secara kontinu.
